```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from palmerpenguins import load_penguins

penguins = load_penguins()
```

```python
penguins.head()
```

```python
fig, ax = plt.subplots()

colors = {'Adelie': 'darkorange', 'Gentoo': 'darkcyan', 'Chinstrap': 'purple'}

# Scatter plot for each species
for species, color in colors.items():
    species_data = penguins[penguins['species'] == species]
    ax.scatter(species_data["flipper_length_mm"], species_data["bill_length_mm"], c=color, alpha=0.5, label=species)

ax.set_xlabel(r'Flipper length (mm)', fontsize=10)
ax.set_ylabel(r'Bill length (mm)', fontsize=10)
ax.set_title('Flipper and bill length')

# Add legend
ax.legend()

ax.grid(True)
fig.tight_layout()

plt.show()
```
