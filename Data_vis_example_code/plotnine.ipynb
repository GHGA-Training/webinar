```python
from plotnine import ggplot, geom_point, aes, stat_smooth, facet_wrap, scale_color_manual, labs, theme
from palmerpenguins import load_penguins
penguins = load_penguins()
```

```python

penguins.head()
```

```python
( ggplot(penguins, aes("bill_length_mm", "flipper_length_mm", color="factor(species)"))
 + geom_point()
 + stat_smooth(method="lm")
 + facet_wrap("~species") 
 + theme(figure_size=(7, 3))
 )
```

```python
( 
    ggplot(penguins, aes(x = "flipper_length_mm", y = "bill_length_mm")) +
    geom_point(aes(color = "species", shape = "species"), size = 3, alpha = 0.8) +
    scale_color_manual(values = ("darkorange","purple","#FFFFFF")) +
    labs(title = "Flipper and bill length",
       subtitle = "Dimensions for Adelie, Chinstrap and Gentoo Penguins at Palmer Station LTER",
       x = "Flipper length (mm)",
       y = "Bill length (mm)",
       color = "Penguin species",
       shape = "Penguin species")
) 
```

```python
import matplotlib.pyplot as plt
import numpy as np

import matplotlib as mpl
from matplotlib.colors import LinearSegmentedColormap, ListedColormap

viridis = mpl.colormaps['viridis'].resampled(3)
```
